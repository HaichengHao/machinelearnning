# 关系图

### 学习本节可以和桑基图对比学习

````python
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
import pyecharts.options as opts
from pyecharts.charts import Graph
````

### 数据准备

#### 接受这个设定:nodes是一个列表,列表里是各个节点的名称字典

````python
nodes = [
    {'name':'node1','symbolSize':10}, #symbolSize指定的是关系图节点的大小
    {'name':'node2','symbolSize':20},
    {'name':'node3','symbolSize':30},
    {'name':'node4','symbolSize':10},
    {'name':'node5','symbolSize':20},
    {'name':'node6','symbolSize':20},
    {'name':'node7','symbolSize':30},
    {'name':'node8','symbolSize':10},
]
links = []
for i in nodes:
    for j in nodes:
        links.append({'source':i.get('name'),'target':j.get('name')})
        
````

### 再接收这个设定,link是由节点关系字典组成的列表,既然要存储的是关系,就有源节点和目标节点

````python
links
````

### 绘制关系图

````python
c = (
    Graph()
    .add(
        '',
        nodes=nodes, #传入节点信息列表
        links=links, # 传入关系信息列表
        repulsion=800, #设置排斥力度,其实就是控制节点之间的距离,
        
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='关系图'
        )
    )
)
````

````python
c.load_javascript()
````

````python
c.render_notebook()
````

## 关系图2

````python
node_data = [
    opts.GraphNode(
        name='node1',
        symbol_size=30
        
    ),
    opts.GraphNode(
        name='node2',
        symbol_size=10
        
    ),
    opts.GraphNode(
        name='node3',
        symbol_size=20
        
    ),
    opts.GraphNode(
        name='node4',
        symbol_size=10
        
    ),
   opts.GraphNode(
        name='node5',
        symbol_size=10
        
    ),

    opts.GraphNode(
        name='node6',
        symbol_size=20
        
    ),
]

link_data =[
    opts.GraphGLLink(
        source = 'node1',
        target='node2',
        value=2
    ),
    opts.GraphGLLink(
        source = 'node2',
        target='node3',
        value=2
    ),
    opts.GraphGLLink(
        source = 'node3',
        target='node4',
        value=2
    ),
    opts.GraphGLLink(
        source = 'node4',
        target='node5',
        value=2
    ),
    opts.GraphGLLink(
        source = 'node5',
        target='node6',
        value=2
    ),
    opts.GraphGLLink(
        source = 'node6',
        target='node1',
        value=2
    ),

]
````

````python
c2 = (
    Graph()
    .add(
        '',
        nodes=node_data, #传入节点信息列表
        links=link_data, # 传入关系信息列表
        repulsion=800, #设置排斥力度,其实就是控制节点之间的距离
        label_opts=opts.LabelOpts(
            position='top'
        ),
        edge_label=opts.LabelOpts(
            is_show=True,
            position='middle',
            formatter='{b}的数据是:{c}'
            
        )
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='关系图'
        )
    )
)
````

````python
c2.load_javascript()
````

````python
c2.render_notebook()
````

### 数据准备

````python
import json
````

### 读取json数据

````python
with open('../others/les-miserables.json','r',encoding='utf-8') as f:
    j = json.load(f) #反序列化(字典化,将json字符串转换为字典形式)
    nodes = j['nodes']
    links = j['links']
    categories = j['categories']
````

````python
j
````

````python
categories
````

### 开始绘图

````python
c3 = (
    Graph(
        init_opts=opts.InitOpts(
            width='800px',
            height='600px'
        )
    )
    .add(
        '',
        nodes=nodes,
        links=links,
        categories=categories,# 关系图节点分类的类目列表
        layout='circular', #设置布局方式
        # edge_label=opts.LabelOpts(
        #     formatter='{a}:{b}:{c}'
        # )
        is_rotate_label=True,#设置标记文字旋转
        linestyle_opts=opts.LineStyleOpts(
            color='source',#设置线条的颜色为源节点的颜色
            curve=0.3, #设置节点连线的弯曲程度
        )
    )
)
````

````python
c3.load_javascript()
````

````python
c3.render_notebook()
````

def add(
    # 系列名称，用于 tooltip 的显示，legend 的图例筛选。
    series_name: str,

    # 关系图节点数据项列表，参考 `opts.GraphNode`
    nodes: Sequence[Union[opts.GraphNode, ,

    # 关系图节点间关系数据项列表，参考 `opts.GraphLink`
    links: Sequence[Union[opts.GraphLink, ,

    # 关系图节点分类的类目列表，参考 `opts.GraphCategory`
    categories: Union[Sequence[Union[opts.GraphCategory, , None] = None,

    # 是否选中图例。
    is_selected: bool = True,

    # 是否在鼠标移到节点上的时候突出显示节点以及节点的边和邻接节点。
    is_focusnode: bool = True,

    # 是否开启鼠标缩放和平移漫游。
    is_roam: bool = True,

    # 节点是否可拖拽，只在使用力引导布局的时候有用。
    is_draggable: bool = False,

    # 是否旋转标签，默认不旋转。
    is_rotate_label: bool = False,

    # 图的布局。可选：
    # 'none' 不采用任何布局，使用节点中提供的 x， y 作为节点的位置。
    # 'circular' 采用环形布局。
    # 'force' 采用力引导布局。
    layout: str = "force",

    # 关系图节点标记的图形。
    # ECharts 提供的标记类型包括 'circle', 'rect', 'roundRect', 'triangle', 
    # 'diamond', 'pin', 'arrow', 'none'
    # 可以通过 'image://url' 设置为图片，其中 URL 为图片的链接，或者 dataURI。
    symbol: Optional[str] = None,

    # 关系图节点标记的大小
    # 可以设置成诸如 10 这样单一的数字
    # 也可以用数组分开表示宽和高，例如 [20, 10] 表示标记宽为20，高为10。
    symbol_size: types.Numeric = 10,

    # 边的两个节点之间的距离，这个距离也会受 repulsion。
    # 支持设置成数组表达边长的范围，此时不同大小的值会线性映射到不同的长度。值越小则长度越长。
    edge_length: Numeric = 50,

    # 节点受到的向中心的引力因子。该值越大节点越往中心点靠拢。
    gravity: Numeric = 0.2,

    # 节点之间的斥力因子。
    # 支持设置成数组表达斥力的范围，此时不同大小的值会线性映射到不同的斥力。值越大则斥力越大
    repulsion: Numeric = 50,

     # Graph 图节点边的 Label 配置（即在边上显示数据或标注的配置）
    edge_label: types.Label = None,

    # 边两端的标记类型，可以是一个数组分别指定两端，也可以是单个统一指定。
    # 默认不显示标记，常见的可以设置为箭头，如下：edgeSymbol: ['circle', 'arrow']
    edge_symbol: Optional[str] = None,

    # 边两端的标记大小，可以是一个数组分别指定两端，也可以是单个统一指定。
    edge_symbol_size: Numeric = 10,

    # 标签配置项，参考 `series_options.LabelOpts`
    label_opts: Union[opts.LabelOpts, dict] = opts.LabelOpts(),

    # 关系边的公用线条样式。
    linestyle_opts: Union[opts.LineStyleOpts, dict] = opts.LineStyleOpts(),

    # 提示框组件配置项，参考 `series_options.TooltipOpts`
    tooltip_opts: Union[opts.TooltipOpts, dict, None] = None,

    # 图元样式配置项，参考 `series_options.ItemStyleOpts`
    itemstyle_opts: Union[opts.ItemStyleOpts, dict, None] = None,
)

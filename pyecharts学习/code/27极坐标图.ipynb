# 极坐标图

```python
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
import pyecharts.options as opts
from pyecharts.charts import Polar
from pyecharts.faker import Faker
import random
```

## 数据准备

```python
data = [(i,random.randint(1,100)) for i in range(101)]
data
```

```python
c = (
    Polar()
    .add(
        '极坐标',
        data,
        type_='scatter',#散点图
        
    )
)
```

```python
c.load_javascript()
```

```python
c.render_notebook()
```

注意极坐标系不是x轴y轴那样的形式

```python
c1 = (
    Polar()
    .add_schema(
        radiusaxis_opts=opts.RadiusAxisOpts(
            data = Faker.week,
            type_='category',
        
        ),
        angleaxis_opts=opts.AngleAxisOpts(
            is_clockwise=True,
            max_=10,
        )
    )
    .add(
        '商品',
        [1,2,3,4,5,3,1],
        type_='bar'
    )

)
```

```python
c1.load_javascript()
```

```python
c1.render_notebook()
```

```python
c2 = (
    Polar()
    .add_schema(
        radiusaxis_opts=opts.RadiusAxisOpts(
            data = Faker.week,
            type_='category',
        
        ),
        # angleaxis_opts=opts.AngleAxisOpts(
        #     data = Faker.week,
        #     type_='category'
        #     # is_clockwise=True,
        #     # max_=10,
        # )
    )
    .add(
        '商品A',
        [1,2,3,4,5,3,1],
        type_='bar',
        stack='abc'
        
    )
    .add(
        '商品B',
        [1,2,3,3,6,7,9],
        type_='bar',
        stack='abc'
        
    )
    .set_global_opts(
        # 设置堆叠柱形图
        title_opts=opts.TitleOpts(
            title='堆叠柱形图',
            
        ),
    )

)
```

```python
c2.render_notebook()
```

```python
c3 = (
    Polar()
    .add_schema(
        # radiusaxis_opts=opts.RadiusAxisOpts(
        #     data = Faker.week,
        #     type_='category',
        
        # ),
        angleaxis_opts=opts.AngleAxisOpts(
            data = Faker.week,
            type_='category'
        #     # is_clockwise=True,
        #     # max_=10,
        )
    )
    .add(
        '商品A',
        [1,2,3,4,5,3,1],
        type_='bar',
        stack='abc'
        
    )
    .add(
        '商品B',
        [1,2,3,3,6,7,9],
        type_='bar',
        stack='abc'
        
    )
    .add(
        '商品C',
        [2,2,5,8,4,1,7],
        type_='bar',
        stack='abc'
        
    )
    .set_global_opts(
        # 设置堆叠柱形图
        title_opts=opts.TitleOpts(
            title='堆叠柱形图',
            
        ),
    )

)
```

```python
c3.render_notebook()
```

```python
data_new = [(i,random.randint(1,100)) for i in range(10)]
```

```python
c4 =(
    Polar()
    .add('涟漪散点极坐标图',data_new,type_='effectScatter',
        effect_opts=opts.EffectOpts( #设置涟漪效果
            scale=10, #设置涟漪范围大小
            period=10 #设置涟漪速度
        ))
)
```

```python
c4.load_javascript()
```

```python
c4.render_notebook()
```

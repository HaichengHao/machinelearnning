# 配置项的使用

<p class='imp'>在pyecharts中,一切皆配置</p>

````python
# 先导入能够解决render_notebook()失效的包
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
````

### 导入配置需要的包

````python
from pyecharts import options as opts #导入配置项
````

````python
from pyecharts.charts import Bar
# 创建bar对象
bar = (
    Bar()
    .add_xaxis(['张三','李四','王五','赵六'])
    .add_yaxis('年龄分布',[18,11,29,33])

    # 设置全局配置
    # .set_global_opts(title_opts=opts.TitleOpts(title='主标题',subtitle='副标题'))
    # 或者
    .set_global_opts(title_opts={'text':'主标题','subtext':'副标题'})

    
)
bar.load_javascript() #首次运行需要加载javascript
````

````python
bar.render_notebook()
````

## 渲染成图片文件


````python
from pyecharts.render import make_snapshot
````

````python
make_snapshot(snapshot,bar.render(),'../others/demopic1.png')
````

## 使用主题

````python
from pyecharts.globals import ThemeType
````

````python
bar=(
    Bar(
        init_opts=opts.InitOpts(theme=ThemeType.DARK) #使用暗色主题
    )
    .add_xaxis(['张三','李四','王五','赵六'])
    .add_yaxis('年龄分布',[18,11,29,33])

    # 设置全局配置
    .set_global_opts(title_opts={'text':'主标题','subtext':'副标题'})
)
bar.render_notebook()
````

````python
bar = (
    Bar(
        init_opts=opts.InitOpts(theme=ThemeType.LIGHT)
    )
    .add_xaxis(['张三','李四','王五','赵六'])
    .add_yaxis('年龄分布A',[18,11,29,33])
    .add_yaxis('年龄分布B',[20,21,28,35])

    # 设置全局配置
    .set_global_opts(title_opts={'text':'主标题','subtext':'副标题'})
)
bar.render_notebook()
````

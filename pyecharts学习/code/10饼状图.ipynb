# 饼状图

```python
import pyecharts.options as opts
from pyecharts.globals import CurrentConfig,NotebookType,RenderSepType,ThemeType
from pyecharts.faker import Faker
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.charts import Pie
```

### 数据准备

```python
data = list(zip(Faker.choose(),Faker.values())) #利用zip将两个列表组成元组
```

## 创建图表

```python
c=(
    Pie()
    .add('',[list(x) for x in data])
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='手机市场',
            
        ),
        # 图例配置
        legend_opts=opts.LegendOpts(
            type_='scroll',
            # 设置图例图标
            legend_icon='circle'
        ),
    )
)
```

```python
c.load_javascript()
```

```python
c.render_notebook()
```

```python
c1 = (
    Pie()
    # 设置颜色,注意这里就和之前画的一些图不一样了
    .set_colors(['red','blue','gray','orange','pink','purple','aqua'])
    .add('手机类型与产值',[list(x) for x in data])
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='手机市场',
            
        ),
        # 图例配置
        legend_opts=opts.LegendOpts(
            type_='scroll',
            # 设置图例图标
            legend_icon='circle',
            pos_left='80%',
            orient='virtical',
            pos_top='30%'
        ),
        # 工具箱配置
        toolbox_opts=opts.ToolboxOpts(
            
        ),
        # 提示框配置
        tooltip_opts=opts.TooltipOpts(
            formatter='{a}:{b}-{c}',
            background_color='orange',
            textstyle_opts=opts.TextStyleOpts(
                color='white'
            )
        ),
    )
    # 设置系列配置项
    .set_series_opts(
        # # 配置图元
        # itemstyle_opts=opts.ItemStyleOpts(
        #     border_color='green',
        #     border_width=5
        # ),
        # # 配置行样式
        # linestyle_opts=opts.LineStyleOpts(
        #     color='blue',
        #     type_='dashed'
        # ),
        # 配置标签
        label_opts=opts.LabelOpts(
            # 设置标签显示的格式,这样设置可以直接显示,关于{a},{b},{c}它们都和tooltipoption中的含义是一样的
           formatter= '{b}:{c}',
        )
    )
)
c1.render_notebook()
```

# 象形柱状图

```python
import pyecharts.options as opts
from pyecharts.globals import CurrentConfig,NotebookType,ThemeType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.faker import Faker
from pyecharts.commons.utils import JsCode

# 导入画象形柱状图要用到的包
# from pyecharts.globals import SymbolType
from pyecharts.charts import PictorialBar
```

## 准备数据

```python
location = ['山西','山东','河南','河北','上海','内蒙古','新疆','西藏']
```

```python
import random
```

```python
values=[]
for i in range(len(location)):
    values.append(random.randint(100,200))
```

```python
values
```

## 创建象形柱状图对象

```python
color_func = """
    function(params){
        if( params >100 && params <140){
            return 'red';
        }
        else if(params > 140 && params< 160){
            return 'tomato';
        }
        else{
            return 'blueviolet';
        }
    }
"""
c = (
    PictorialBar(
        init_opts=opts.InitOpts(
            # theme=ThemeType.ESSOS
        )
    )
    .add_xaxis(location)
    # 在添加y轴中设置象形符号
    .add_yaxis('',values,label_opts=opts.LabelOpts(is_show=False),
               symbol_size=16,#设置符号大小
               symbol='round_rect', #设置符号类型
               symbol_repeat='fixed', # 设置符号重复
               is_symbol_clip=True, #是否进行裁剪,需要设置为True
               itemstyle_opts=opts.ItemStyleOpts(
        color=JsCode(color_func)
    ))
    .reversal_axis()
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            name='数值',
            name_location='end',
            is_show=False
        ),
        yaxis_opts=opts.AxisOpts(
            name='省份',
            name_location='end',
            axistick_opts=opts.AxisTickOpts(
                is_show=False
            ),
            axisline_opts=opts.AxisLineOpts(
                is_show=False
            )
            
        ),
        title_opts=opts.TitleOpts(
            title='主标题',
            subtitle='副标题'
        ),
        legend_opts=opts.LegendOpts(
            # type_='scroll',
            legend_icon='triangle'
            ,inactive_color='black'
        ),
        toolbox_opts=opts.ToolboxOpts(),
        tooltip_opts=opts.TooltipOpts(
            background_color='orange',
            border_color='purple',
            border_width=2,
            textstyle_opts=opts.TextStyleOpts(
                color='white'
            )
        ),
        visualmap_opts=opts.VisualMapOpts(
            range_text=['高','低']
        )
    )
    # .set_series_opts()
    
)
```

```python
c.load_javascript()
```

```python
c.render_notebook()
```

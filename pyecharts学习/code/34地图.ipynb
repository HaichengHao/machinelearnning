# 地图

```python
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
import pyecharts.options as opts
from pyecharts.charts import Map
from pyecharts.faker import Faker
```

### 绘制地图

#### 数据准备,利用Faker获取省份信息

```python
Faker.provinces
```

```python
c = (
    Map()
    .add(
        '',
        [list(i) for i in zip(Faker.provinces,Faker.values())],#准备的数据应当是由省份和数值组成的一个列表
        maptype='china' #地图类型,指定我国
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='地图'
        ),
        visualmap_opts=opts.VisualMapOpts(
            range_text=['高','低'],
            # is_piecewise=True
            max_=150
            
        )
    )
)
```

```python
c.load_javascript()
```

```python
c.render_notebook()
```

## 地球

```python
from pyecharts.faker import POPULATION #注意,目前最新版本已经删除掉了,所以为了复现这个案例,我们需要将pyecharts包降级到1.9.0
from pyecharts.charts import MapGlobe
```

```python
POPULATION
```

#### 我们不需要第列表中索引0的数据,因为那是列表标头,而我们进行可视化是不需要的,所以要去掉,那么拿数据时候我们只需要限定为POPULATION[1:]就行,然后我们也不需要列表元素中的国家的名称元素,所以利用_匹配,再使用列表生成式,将列表元素中第二个元素获取即可

```python
data = [x for _,x in POPULATION[1:]] #利用占位符匹配POPULATION[1:]列表中的列表元素中的人口量数字
```

```python
#定义两个变量分别接收人口数中的最小值和最大值,我们可以用它们来进行视觉映射的划分
low,heigh=min(data),max(data)
```

```python
low,heigh
```

```python
c4=(
    MapGlobe()
    .add_schema()
    .add(
        maptype='world',
        series_name='世界人口',
        data_pair=POPULATION[1:],
        #data_pair参数接收的参数为二元数组，第一个元素为区域名称，第二个为区域值,
        # 我们传入的是除了标头的所有数据列表
        is_map_symbol_show=False, #将地图显示符号指定为False,这个参数是控制各个国家的符号是否显示的
        label_opts=opts.LabelOpts(
            is_show=False
        )
    )
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(
            min_=low,
            max_=heigh,
            range_text=['多','少'],
            is_calculable=True, #
        )
        
    )
    
    
)
```

```python
c4.load_javascript()
```

```python
from pyecharts.charts import Map
# pyecharts自带的世界各国人口数据
from pyecharts.faker import POPULATION


data = [x for _, x in POPULATION[1:]]
low, high = min(data), max(data)
(
    Map(init_opts=opts.InitOpts())
    .add("World Pupulation", POPULATION[1:], "world", is_map_symbol_show=False)
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(visualmap_opts=opts.VisualMapOpts(max_=high, min_=low))
    .render_notebook()
)
```

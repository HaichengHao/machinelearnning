# 坐标轴配置项

```python
from pyecharts import options as opts
from pyecharts.globals import CurrentConfig,NotebookType,ThemeType,RenderType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.faker import Faker
from pyecharts.charts import Bar
```

```python
c=(
    Bar(
        init_opts=opts.InitOpts(
            width='800px',
            height='750px',
            # theme=ThemeType.ESSOS,
            # bg_color='bisque'
            
        )
    )
    .add_xaxis(Faker.choose())
    .add_yaxis('A',Faker.values())
    .add_yaxis('B',Faker.values())
    .add_yaxis('C',Faker.values())
    .add_yaxis('D',Faker.values())
    
    .set_global_opts(
        # 设置标题配置项
        # TitleOptions
        title_opts=opts.TitleOpts(
            title='主标题',
            subtitle='副标题',
            # pos_bottom=0,
            # pos_left=0,
            # padding=10, #注意padding直接写就可以不需要用引号引起来
            # item_gap= 40 #设置主标题和副标题的间距
            # 写法很多,也可以pos_方向='npx'这种形式指定像素
            # 也可以写pos_方向='n%'这样的形式，结合前端知识理解更快更容易
            # 我们也可以写内边距不得不佩服baidu前端大佬写的echarts对与有前端基础的人太友好了
            
        ),
        # 设置图例配置项
        legend_opts=opts.LegendOpts(
            # 图例的类型有两种 ,默认是plain,还有一种是可以滚动scroll,适合图例非常多的情况
            type_="scroll",
            orient='vertical',#控制方向,两个选择,水平或者垂直
            pos_right=0, #像定位这些很多全局配置项目都是通用的
            pos_top=30,
            # is_show=False #是否显示图例,如果是False则会隐藏图例
            # 选择模式
            # 4个值,True为开启图例点击(默认),False为不开启图例点击,这俩不用加引号
            # single：默认单选 multipel:多选 这俩要加引号
            selected_mode='multiple',
            # item_gap= 3 #没错这个和上面的titleopts也是差不多的,不过控制的是图例之间的间距
            # align='right', #默认是auto,设置图例的对齐方式,相对于文字而言的色块儿的位置
            # item_width=10, #设置图例的宽度 
            # item_height=10, #设置图例的高度
            inactive_color='black', #设置图例非激活状态下的颜色默认是灰色
            legend_icon= 'circle'  #甚至可以改图例的图表样式
            # 常见的图标 'circle','rect','roundRect(圆角矩形，默认选项)','triangle(三角形)','dismond','arrow'
            # 甚至你可以自己指定一个图标的文件路径也行,就像是html中可以指定cursour即鼠标的样式
        ),
        # 设置工具箱配置ToolboxOpts
        toolbox_opts=opts.ToolboxOpts(),
        # 设置视觉映射
        visualmap_opts=opts.VisualMapOpts(
            type_='color', #两个选项,一个是color,一个是size
            min_=30, #设置最小值
            max_=90 ,#设置最大值
            range_color=['red','green','blue','purple'], #指定视觉映射组件的过渡颜色
            range_opacity=0.7, #图元和标签文字都透明
            # range_start=,
            range_text=['大','小'], #设置的是视觉映射最大值和最小值的文本
            # range_size=
            orient='vertical' ,#设置对齐方式和之前都是一样的，默认是垂直对齐
            # pos_right=,
            # pos_top=,
            # is_piecewise=True #是否为分段型,默认是False
            # is_inverse= #是否将组件翻转,默认是False
        ),
        # 设置提示框配置项
        tooltip_opts=opts.TooltipOpts(
            # is_show=True #默认显示
            # 触发类型
            # 有几个值可供选择
            # item:数据项,一般用于散点图,柱状图,饼图(默认项)
            # axis:坐标轴,提示线,主要用于条形图,折线图
            # trigger='axis'

            # 触发条件
            # 几个选项可供选择
            # mousemove(默认),mousemove,click,也可以这样写mousehover|click采用双约束触发
            trigger_on='mousemove|click',
            is_show_content=True, #是否显示提示框的内容,默认是True

            # 标签内容的格式
            # 几个常用的模板变量
            # {a}:系列名,就是add_yaxis中的参数中的系列名,即A类,B类这样的
            # {b}:数据名
            # {c}:值
            formatter='{a}:{b}--{c}',
            background_color='tomato', #提示框设置背景颜色,注意这里与初始化配置中的写法略微不同
            border_color='white', #设置边框的颜色,注意如果不设置border_width可能会看不出看来
            border_width=2,
            textstyle_opts=opts.TextStyleOpts(
                color='green' #设置提示框中字体的颜色
            )
        ),
        # 设置数据缩放配置项
        # DatazoomOptions
        datazoom_opts=opts.DataZoomOpts(
            is_show=True, #是否显示底部滑块组件
            # type_='slide', #设置滑块组件样式,默认就是slide还有一个是inside
            # 我们来看一下inside
            # type_='inside',
            is_realtime=True, #是否实时渲染图表默认True
            range_start=20,#设置滑块区域(数据窗口的起始位置)的默认位置,默认是20，我们可以自己指定
            orient='horizontal', #修改水平方向默认是horizontal,如果改成垂直那就是virtical
            # is_zoom_lock=True #是否锁定选择区域,如果锁定那就不能缩放只能平移
        ),
        # AxisOpts: 坐标配置项
        xaxis_opts=opts.AxisOpts(
            is_show=True, #是否显示x轴,默认是True
            #坐标轴的类型
            # 有几个可供选择的值
            # value:数值轴 , 一般用于连续型数据
            # category:类目轴,适用于离散数据,比如星期一到星期五这样的
            # time:时间轴,适用于连续的时序数据
            type_='category',
        ),
        yaxis_opts=opts.AxisOpts(
            axisline_opts=opts.AxisLineOpts(
                # is_show=False, #不显示y轴的竖线
            ),
            axistick_opts=opts.AxisTickOpts(
                # is_show=False,#不显示y轴的刻度
            )
        )
    )
)
```

# 多图布局

## 需要一个新引入的包Grid

```python
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
import pyecharts.options as opts
from pyecharts.faker import Faker
from pyecharts.charts import Grid,Line,Scatter,EffectScatter,Bar
```

#### 数据准备

```python
x_data = Faker.clothes
y_data = Faker.values()
```

## 折线图

```python
line = (
    Line()
    .add_xaxis(x_data)
    .add_yaxis('A',y_data,is_smooth=True,areastyle_opts=opts.AreaStyleOpts(opacity=0.7),
              label_opts=opts.LabelOpts(
                  rotate=20
              ),
              symbol='triangle',
              symbol_size=20,
               
            )
    .add_yaxis('B',Faker.values(),is_smooth=True,areastyle_opts=opts.AreaStyleOpts(opacity=0.7),
              label_opts=opts.LabelOpts(
                  rotate=20
              ),
              symbol='dotted',
              symbol_size=20,
               
            )
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            boundary_gap=False #去掉x起始位置与y轴的间距,可以让面积图看起来更美观
        ),
        legend_opts=opts.LegendOpts(
            
        )
    )
)
```

```python
line.load_javascript()
```

```python
line.render_notebook()
```

## 散点图

```python
scatter = (
    EffectScatter()
    .add_xaxis(x_data,)
    .add_yaxis('A',y_data,symbol='dotted',color='blue')
    .add_yaxis('B',Faker.values(),symbol='arrow',color='red')
    .set_global_opts(
        legend_opts=opts.LegendOpts(
            pos_right='20%'
        )
    )
    
)
```

```python
scatter.load_javascript()
```

```python
scatter.render_notebook()
```

### 接下来是多图布局

## 多图布局

其实很简单,相当于创建一个Grid对象,然后把要进行多图布局的图案放到这个容器里边就行

#### 多图布局采用opts.GridOpts控制布局

```python
Grid=(
    Grid()
    .add(
        line,
        grid_opts=opts.GridOpts(
            pos_right='55%'
        )
    )
    .add(
        scatter,
        grid_opts=opts.GridOpts(
            pos_left='55%'
        )
    )
)
```

```python
Grid.load_javascript()
```

```python
Grid.render_notebook()
```

### 总体写法

```python
from pyecharts.charts import Grid
line2 = (
    Line()
    .add_xaxis(x_data)
    .add_yaxis('A',y_data,is_smooth=True,areastyle_opts=opts.AreaStyleOpts(opacity=0.7),
              label_opts=opts.LabelOpts(
                  rotate=20
              ),
              symbol='triangle',
              symbol_size=20,
               
            )
    .add_yaxis('B',Faker.values(),is_smooth=True,areastyle_opts=opts.AreaStyleOpts(opacity=0.7),
              label_opts=opts.LabelOpts(
                  rotate=20
              ),
              symbol='dotted',
              symbol_size=20,
               
            )
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            boundary_gap=False #去掉x起始位置与y轴的间距,可以让面积图看起来更美观
        ),
        legend_opts=opts.LegendOpts(
            pos_left='20%'
        )
    )
)
scatter2 = (
    EffectScatter()
    .add_xaxis(x_data,)
    .add_yaxis('c',y_data,symbol='dotted',color='blue')
    .add_yaxis('d',Faker.values(),symbol='arrow',color='red')
    .set_global_opts(
        legend_opts=opts.LegendOpts(
            pos_right='20%'
        )
    )
    
)
grid=(
    Grid()
    .add(
        line2,
        grid_opts=opts.GridOpts(
            pos_right='55%'
        )
    )
    .add(
        scatter2,
        grid_opts=opts.GridOpts(
            pos_left='55%'
        )
    )
)
```

```python
grid.load_javascript()
```

```python
grid.render_notebook()
```

## 重叠图overlap()

### 数据准备

```python
a = Faker.cars
b = Faker.values()
a,b
```

```python
bar = (
    Bar()
    # .add_xaxis(a) #注意,组合图中留一个x就可以,line2的有x轴了,没必要再来一个同样的,不然就会显示三组数据
    .add_yaxis('种类',b)
)
```

```python
line2 = (
    Line()
    .add_xaxis(a)
    .add_yaxis('',b,symbol='arrow')
)
```

```python
bar.load_javascript,line2.load_javascript()
```

```python
mix = line2.overlap(bar)
```

```python
mix.set_global_opts(
    visualmap_opts=opts.VisualMapOpts(
        min_=20,
        max_=130,
        range_color=Faker.visual_color
    ),
    title_opts=opts.TitleOpts(
        title='组合图'
    ),
    toolbox_opts=opts.ToolboxOpts(),
    legend_opts=opts.LegendOpts(

        inactive_color='black'
    ),
    
)
```

```python
mix.render_notebook()
```

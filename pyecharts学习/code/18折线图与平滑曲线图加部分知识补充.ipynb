```python
import pyecharts.options as opts
from pyecharts.globals import CurrentConfig,NotebookType,ThemeType
from pyecharts.charts import Line
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.faker import Faker
```

### 数据准备

```python
v = Faker.clothes
v
```

```python
y1_v = Faker.values()
y1_v
```

# 折线图

```python
c = (
    Line(
        init_opts=opts.InitOpts(
            # theme=ThemeType.ESSOS
        )
    )
    .add_xaxis(v)
    .add_yaxis('价格',y1_v)
    
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            name = '衣物',
            name_location='end'
        ),
        yaxis_opts=opts.AxisOpts(
            name='价格(元)',
            name_location='end'
        ),
        title_opts=opts.TitleOpts(
            title='服装价格',
            # subtitle=''
        ),
        tooltip_opts=opts.TooltipOpts(
           
        ),
        toolbox_opts=opts.ToolboxOpts(),
        visualmap_opts=opts.VisualMapOpts(),
        datazoom_opts=opts.DataZoomOpts()
        
        
    )
    .set_series_opts(
        linestyle_opts=opts.LineStyleOpts(
            # color='orange'
        ),
        itemstyle_opts=opts.ItemStyleOpts(
            # color='red',
        ),
        markpoint_opts=opts.MarkPointOpts(
            data=[
                opts.MarkPointItem(
                    type_='max'
                )
            ]
        )
    )
    
    
)
```

```python
c.load_javascript()
```

```python
c.render_notebook()
```

# 平滑曲线图

其实只需要在.add_yaxis的时候加上一个<span class='imp'> is_smooth=True</span>即可

```python
# 如果我们想要设置平滑曲线,则需要设置is_smooth为True,这样就可以实现matplotlib中的plot效果
c1 = (
    Line(
        init_opts=opts.InitOpts(
            # theme=ThemeType.ESSOS
            # width=,
            # type_='dashed'
        )
    )
    .add_xaxis(v)
    .add_yaxis('新价格',Faker.values())

    # 注意,对之前没写全的内容进行补充
    # 这里补上的内容有我们之前一直没有写的series_name(系列名)以及y_axis(y轴数据),symbol(符号点的样式)
    .add_yaxis(series_name='旧价格',y_axis=y1_v,
               symbol='triangle',symbol_size=10,#这里只需要加上symbol即可修改成自己想要的图元样式,而不需要单独再去系列配置里写了
               is_smooth=True,
               linestyle_opts=opts.LineStyleOpts(type_='dashed'),
               itemstyle_opts=opts.ItemStyleOpts(
                   border_color='purple',
                   border_width='4'
               )
               )
    
    .set_global_opts(
        # graphic_opts=opts.Gra
        xaxis_opts=opts.AxisOpts(
            name = '衣物',
            name_location='end',
            axistick_opts=opts.AxisTickOpts(
                linestyle_opts=opts.LineStyleOpts(
                    color='blue'
                )
            ),
            axisline_opts=opts.AxisLineOpts(
                linestyle_opts=opts.LineStyleOpts(
                    color='red',
                )
            ),
            
        ),
        yaxis_opts=opts.AxisOpts(
            name='价格(元)',
            name_location='end',
            splitline_opts=opts.SplitLineOpts(
                is_show=True
            )
        ),
        title_opts=opts.TitleOpts(
            title='服装价格',
            # subtitle=''
        ),
        legend_opts=opts.LegendOpts(
            # item_width=,
            # legend_icon='circle'
        ),
        tooltip_opts=opts.TooltipOpts(
           trigger='axis'
        ),
        toolbox_opts=opts.ToolboxOpts(),
        visualmap_opts=opts.VisualMapOpts(),
        datazoom_opts=opts.DataZoomOpts()
        
        
    )
    # 这里还要强调一点,那就是系列配置也是可以在添加y轴数据的时候写入的
    .set_series_opts(
        # label_opts=opts.LabelOpts(
        #     is_show=False
        # ),
        linestyle_opts=opts.LineStyleOpts(
            # color='orange'
        ),
        # 控制图元样式
        # itemstyle_opts=opts.ItemStyleOpts(
        #     # color='red',
            
        # ),
        markpoint_opts=opts.MarkPointOpts(
            data=[
                opts.MarkPointItem(
                    type_='max'
                )
            ]
        ),
        markline_opts=opts.MarkLineOpts(
            data=[
                opts.MarkLineItem(
                    type_='average',
                    
                )
            ]
        )
    )
    
    
)
```

```python
c1.load_javascript()
```

```python
c1.render_notebook()
```

## 导入需要的包

```python
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.faker import Faker
from pyecharts.globals import ThemeType,RenderType
```

### 全局配置项

- 全局配置可通过set_global_opts方法设置

```python
bar = (
    Bar(
        #设置主题
        init_opts=opts.InitOpts(theme=ThemeType.ROMA)
    )
    .set_global_opts(
        # TitleOpts: 标题配置项
        title_opts=opts.TitleOpts(
            title='主标题',
            subtitle='副标题',
            # 设置标题链接
            title_link='https://www.being.com',
            subtitle_link='https://www.badidu.com',
            title_target='blank',#这里就和html之中链接的写法参数差不多,写blank就表示在新窗口打开,写self就在当前的页面中打开
            # 默认就是blank

            # 控制标题的位置(共四个方向,left,right,bottom,top)
            # 让标题靠下靠右显示,结合前端知识理解
            pos_bottom=0,
            pos_right=0,
            # 写法很多,也可以pos_方向='npx'这种形式指定像素
            # 也可以写pos_方向='n%'这样的形式，结合前端知识理解更快更容易
            # 我们也可以写内边距不得不佩服baidu前端大佬写的echarts对与有前端基础的人太友好了
            padding=10,
            item_gap=40 #设置主标题和副标题之间的间距
        ),
        # title_opts={'text':'主标题','subtext':'副标题'}
        # LegendOpts:图例配置项
        legend_opts=opts.LegendOpts(),
        # ToolboxOpts:工具箱配置项
        toolbox_opts=opts.ToolboxOpts(),
        # VisualMapOpts:视觉映射配置项
        visualmap_opts=opts.VisualMapOpts(),
        #TooltipOpts:提示框配置项
        tooltip_opts=opts.TooltipOpts(),
        # DataZoomOpts:区域缩放配置项
        datazoom_opts=opts.DataZoomOpts(
            is_show=True, #是否显示底部滑块组件
            # type_='slide', #设置滑块组件样式,默认就是slide还有一个是inside
            # 我们来看一下inside
            # type_='inside'
            is_realtime=True, #是否实时渲染图表默认True
            range_start=20,#设置滑块区域(数据窗口的起始位置)的默认位置,默认是20，我们可以自己指定
        )
    )
    .add_xaxis(['张三','李四','王五','赵六'])
    .add_yaxis(series_name='年龄分布',y_axis=[18,11,29,33])
)
```

```python
bar.load_javascript()
```

```python
bar.render_notebook()
```

## 初始化配置项(init_opts)写在Bar()中

```python
c = (
    Bar(
        # 初始化配置项
        init_opts=opts.InitOpts(
            # 指定宽度
            width='800px',
            height='500px',#图表的画布大小,长度单位和css是一样的
            # 设置渲染者
            renderer=RenderType.CANVAS, #一个渲染风格是CANVAS,也可以用SVG
            # 设置网页标题
            page_title='网页标题1',
            # 设置主题
            theme=ThemeType.WONDERLAND,
            # 设置背景颜色
            bg_color='tomato'
        )
    )
    .add_xaxis(Faker.choose())
    .add_yaxis('商家A',Faker.values())
    .add_yaxis('商家B',Faker.values())
    # 设置全局配置
    .set_global_opts(
        title_opts={'text':'货品比价'}
    )
)
c.render_notebook()
```

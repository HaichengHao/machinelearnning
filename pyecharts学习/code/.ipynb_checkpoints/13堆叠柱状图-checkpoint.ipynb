# 堆叠柱状图

```python
import pyecharts.options as opts
from pyecharts.globals import CurrentConfig,NotebookType,ThemeType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.faker import Faker
from pyecharts.charts import Bar
```

```python
a = Faker.choose()
b = Faker.values()
```

```python
a
```

```python
c = (
    Bar(
        init_opts=opts.InitOpts(
            page_title='义乌服饰销售量',
            theme=ThemeType.ESSOS,
            # bg_color=''
        )
    )
    .add_xaxis(a)
    # 如果想要绘制堆叠柱状图只需要在添加y坐标的值的时候给需要堆叠的两组数据指定同一个stack即可
    .add_yaxis('商家A',b,stack='abc') #这里指定abc就是给其指定了一个堆叠栈,如果想要另外一个和其堆叠，只需要给另外一个也加上同样的stack
    .add_yaxis('商家B',Faker.values(),stack='abc')
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='义乌服饰销量',
            subtitle='日销量'
        ),
        # 设置图例
        legend_opts=opts.LegendOpts(
            legend_icon='circle',
            # type_='scroll',
            # pos_right='',
            inactive_color='black'
        ),
        # 设置工具箱
        toolbox_opts=opts.ToolboxOpts(),
        # 设置可视化映射
        visualmap_opts=opts.VisualMapOpts(
            range_text=['多','少']
        ),
        # 设置提示框
        tooltip_opts=opts.TooltipOpts(
            background_color='orange',
            textstyle_opts=opts.TextStyleOpts(
                color='white'
            )
        ),
        # 设置数据缩放
        datazoom_opts=opts.DataZoomOpts(
            # type_='inside', #实现滚动缩放
        )
    )
    # 设置系列配置项
    .set_series_opts(
        # 标签文字
        label_opts=opts.LabelOpts(
            # position='',
            # rotate=0,
        ),
        # 图元设置
        itemstyle_opts=opts.ItemStyleOpts(
            border_color='white',
            border_width=2,
            opacity=0.7, #设置不透明度,0为完全透明,1为完全不透明
        ),
        # 线条配置
        linestyle_opts=opts.LineStyleOpts(),

        # 标记点和标记线配置项
        markpoint_opts=opts.MarkPointOpts(
            data=[
                # 最大值点
                opts.MarkPointItem(type_='max',name='最大值'),
                #最小值点
                opts.MarkAreaItem(type_='min',name='最小值')
            ],
            
        ),
        # 标记线配置
        markline_opts=opts.MarkLineOpts(
            data=[
                opts.MarkLineItem(type_='average',name='平均值线')
            ],
            linestyle_opts=opts.LineStyleOpts(
                type_='dashed',
                # color='red'
            ),
            # 控制标签文字
            label_opts=opts.LabelOpts(
                color='red'
            )
            
            )
        )
    )
```

```python
c.load_javascript()
```

```python
c.render_notebook()
```

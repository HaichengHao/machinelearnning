# 系列配置项值标记线和标记点配置

### 导入需要用到的包

```python
import pyecharts.options as opts
from pyecharts.globals import CurrentConfig,NotebookType,RenderType,ThemeType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.charts import Line
from pyecharts.faker import Faker
```

```python
c = (
    Line(
        # 初始化配置
        init_opts=opts.InitOpts(
            theme=ThemeType.ESSOS,
            page_title='我是一个页面名称'
        )
    )
    .add_xaxis(Faker.choose())
    .add_yaxis('类别A',Faker.values())
    .add_yaxis('类别B',Faker.values())
    .add_yaxis('类别C',Faker.values())
    # 设置全局配置
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='主标题',
            subtitle='副标题',
        ),
        legend_opts=opts.LegendOpts(
            type_='scroll',
            orient='virtical',
            pos_right=0,
            inactive_color='black',
            legend_icon='circle'
            
        ),
        toolbox_opts=opts.ToolboxOpts(
            pos_bottom=150,
            orient='vertical',
            pos_left=0,
        ),
        visualmap_opts=opts.VisualMapOpts(
            # range_color=,
            # orient=,
            # min_=,
            # max_=,
            # is_inverse=,
        ),
        tooltip_opts=opts.TooltipOpts(
            background_color='orange',
            trigger='axis',
            border_color='white',
            border_width=3,
            textstyle_opts=opts.TextStyleOpts(
                color='white',
            )
        ),
        datazoom_opts=opts.DataZoomOpts(
            range_start=20,
            is_zoom_lock=False,
        )
        
    )
    # 设置系列配置项
    .set_series_opts(
        # 标签文字配置项
        label_opts=opts.LabelOpts(
            position='right',
            rotate=0,#默认是0
            # color='red'
        ),
        # 图元配置项,就是折线图上小数据节点的配置,配置它也会影响label标签的颜色
        itemstyle_opts=opts.ItemStyleOpts(
            border_color='red',
            border_width=3,
            opacity=0.7
        ),
        # 线条配置项
        linestyle_opts=opts.LineStyleOpts(
            color='blue',
            type_='dashed'
        ),

        #单独拎出来标记带你配置项和标记线配置项是因为它们的写法略微和前面的系列配置项不同
        
        # 标记点配置项
        markpoint_opts=opts.MarkPointOpts(
            data=[
                # type_ ,特殊的标记类型,有三种max,min,average
                # symbol,符号样式,默认是pin也可以指定成其它的样式,如circle,其实这里不建议修改成其它样式
                # symbolsize,符号尺寸,可以设置markerpoint的大小,同样不建议修改
                opts.MarkPointItem(type_='max',symbol='pin',symbol_size=50),
                # 当然,也可以再来个最小值
                opts.MarkPointItem(type_='min')
            ]
        ),
        # 标记线配置项
        markline_opts=opts.MarkLineOpts(
            data=[
                # 其实标记线中的很多参数和标记点都是差不多的
                # name:给标记起名字
                opts.MarkLineItem(type_='average',name='平均值点')
            ],
            # 控制标记线中的文字的样式
            label_opts=opts.LabelOpts(
                color='red', #控制颜色设置为红色
            )
        )
    )
)
c.load_javascript()
```

```python
c.render_notebook()
```

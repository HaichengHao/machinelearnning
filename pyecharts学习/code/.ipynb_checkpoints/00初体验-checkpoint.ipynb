#### 首先开始绘制第一个图表

````python
from pyecharts.charts import Bar
# 解决render_notebook不显示的方法
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
bar = Bar() #创建一个柱形图对象
````

````python
bar.add_xaxis(['衬衣','羊毛衫','西装','裤子','鞋子','袜子']) #添加x轴,注意是列表
bar.add_yaxis('商家',[5,20,40,100,70,90]) #添加y轴
# 开始渲染,默认会渲染成html
# 在notebook中渲染
bar.load_javascript()
````

````python
bar.render_notebook()
````

````python
# 渲染到指定的文件路径当中
# bar.render('../others/demo1.html') #可以指定生成的路径
````

#### 链式调用

````python
bar=(
    Bar()
    .add_xaxis(['张三','李四','王五','赵六'])
    .add_yaxis('年龄',[18,19,20,22])
    
    
)
````

````python
bar.render_notebook()
````

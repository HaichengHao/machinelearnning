# 3D柱状图

```python
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.faker import Faker
from pyecharts.charts import Bar3D
import random
import pyecharts.options as opts
```

### 数据准备

```python
data = [(i,j,random.randint(0,12)) for i in range(6) for j in range(24)]
```

```python
data
```

### 开始绘图

```python
c = (
    Bar3D()
    .add(
        '系列名称',
        # data, #其实也可以这样写[[d[0],d[1],d[2]] for d in data]
        [[d[1],d[0],d[2]] for d in data],
        xaxis3d_opts=opts.Axis3DOpts(
            Faker.clock,
            # type_='value',
            type_='category',
            name='x轴的名称',
        ),
        yaxis3d_opts=opts.Axis3DOpts(
            Faker.week,
            type_='category',
            name='y轴的名称'
        ),
        zaxis3d_opts=opts.Axis3DOpts(
            type_='value',
            name='z轴名称'
        ),
        grid3d_opts=opts.Grid3DOpts(
            # width=300
            
        )
    )
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(
            min_=0,
            max_=12,
            range_color=Faker.visual_color
        ),
        tooltip_opts=opts.TooltipOpts(
            is_show=True,
            trigger='item'
        )
        
    )
)
```

```python
c.load_javascript()
```

```python
c.render_notebook()
```

### 堆叠柱状图

#### 数据准备

```python
x_data = y_data = list(range(10))
```

```python
x_data,y_data
```

```python
data=[]
for j in range(10):
    for k in range(10):
        value=random.randint(0,9)
        data.append([j,k,value*2+4])
```

```python
data
```

#### 创建一个函数

每次调用这个函数我们就可以得到一个新的数据

```python
def generate_data():
    data=[]
    for j in range(10):
        for k in range(10):
            value=random.randint(0,9)
            data.append([j,k,value*2+4])
    return data
```

```python
generate_data()
```

```python
bar3d = Bar3D()
# 利用循环多生成几组数据产生堆叠柱状图
for _ in range(10):
    bar3d.add(
        '',
        generate_data(),
        shading='lambert',#提升图像的清晰程度,给每个柱子设置阴影效果
        xaxis3d_opts=opts.Axis3DOpts(
            x_data, #设置坐标轴的数据源,其实就是我们之前学过的data=...
            name='x轴的名称',
            type_='value' #连续型数据一般用value,离散型数据一般用category
        ),
        yaxis3d_opts=opts.Axis3DOpts(
            y_data,
            name='y轴的名称',
            type_='value'
        ),
        zaxis3d_opts=opts.Axis3DOpts(
            name='z轴名称',
            type_='value'
        ),
    )
bar3d.set_global_opts(
    title_opts=opts.TitleOpts(
        title='堆叠柱状图',
    ),
    toolbox_opts=opts.ToolboxOpts(),
    tooltip_opts=opts.TooltipOpts()
)
bar3d.set_series_opts(stack='abc')
```

```python
bar3d.load_javascript()
```

```python
bar3d.render_notebook()
```

# 漏斗图

```python
import pyecharts.options as opts
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.faker import Faker
from pyecharts.charts import Funnel #导入漏斗图
```

## 数据准备

```python
lst = [list(i) for i in zip(Faker.clothes,Faker.values())]
```

```python
lst
```

```python
c = (
    Funnel()
    .add(
        '商品', #添加系列名
        [list(i) for i in lst] #准备数据，
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='漏斗图',
            
        ),
        visualmap_opts=opts.VisualMapOpts(),
        tooltip_opts=opts.TooltipOpts(
            trigger='item',
            formatter='{a}<br>{b}:{c}'
        )
    )
)
```

```python
c.load_javascript()
```

```python
c.render_notebook()
```

```python
x_data = Faker.cars
x_data
```

```python
y_data = Faker.values()
y_data
```

```python
data = [ [x_data[i],y_data[i]] for i in range(len(x_data))]
data
```

```python
c2 = (
    Funnel()
    .add('车型',[list(i) for i in data],
        label_opts=opts.LabelOpts(
            is_show=True,
            position='inside'
        ))
    
)
```

```python
c2.render_notebook()
```

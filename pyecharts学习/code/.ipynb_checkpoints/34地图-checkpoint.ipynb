# 地图

```python
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
import pyecharts.options as opts
from pyecharts.charts import Map
from pyecharts.faker import Faker
```

### 绘制地图

#### 数据准备,利用Faker获取省份信息

```python
Faker.provinces
```

```python
c = (
    Map()
    .add(
        '',
        [list(i) for i in zip(Faker.provinces,Faker.values())],#准备的数据应当是由省份和数值组成的一个列表
        maptype='china' #地图类型,指定我国
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='地图'
        ),
        visualmap_opts=opts.VisualMapOpts(
            range_text=['高','低'],
            # is_piecewise=True
            max_=150
            
        )
    )
)
```

```python
c.load_javascript()
```

```python
c.render_notebook()
```

### 显示一个省份的地图

```python
Faker.guangdong_city
```

```python
c2 = (
    Map()
    .add(
        '',
        [list(i) for i in zip(Faker.guangdong_city,Faker.values())],
        maptype='广东' #要进行理解,我们准备的数据是广东省里七个市的数据,那么要将其表示出来,需要广东省的地图作为容器
    ) 
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='地图'
        ),
        visualmap_opts=opts.VisualMapOpts(
            range_text=['高','低'],
            # is_piecewise=True
            max_=150
            
        )
    )
)
```

```python
c2.load_javascript()
```

```python
c2.render_notebook()
```

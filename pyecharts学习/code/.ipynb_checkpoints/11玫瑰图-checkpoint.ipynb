# 玫瑰图

```python
import pyecharts.options as opts
from pyecharts.globals import NotebookType,CurrentConfig,ThemeType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.charts import Pie,Bar
from pyecharts.faker import Faker
```

```python
a = Faker.choose()
b = Faker.values()
```

```python
a
```

```python
b
```

```python
c = (
    Pie()
    .add(
        '犬类',[list(item) for item in zip(a,b)],
        radius=['10%','55%'], #radius控制空心圆半径和外径
        center=['25%','50%'],
        rosetype='radius', #控制玫瑰图的类型

        # 利用LabelOpts控制标签文字的格式
        label_opts=opts.LabelOpts(
            is_show=False
        )
    )
    # 再来一个这次换一下随机生成的数据
    .add(
         '犬类',[list(item) for item in zip(a,Faker.values())],
        radius=['25%','75%'], #radius控制空心圆半径和外径
        center=['65%','50%'], #控制玫瑰图的位置,第一个控制的是水平方向,第二个是垂直方向
        rosetype='radius', #控制玫瑰图的类型

        # 利用LabelOpts控制标签文字的格式
        label_opts=opts.LabelOpts(
            is_show=True
        )
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='玫瑰图'
        )
    )
    
)
```

```python
c.load_javascript()
```

```python
c.render_notebook()
```

## 简单柱状图

```python
import pyecharts.options as opts
from pyecharts.globals import CurrentConfig,NotebookType,ThemeType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.charts import Bar
from pyecharts.faker import Faker

# 添加js要用的包
from pyecharts.commons.utils import JsCode
```

```python
a = Faker.choose()
```

```python
b = Faker.values()
```

```python
c = (
    Bar(
        init_opts=opts.InitOpts(
            theme=ThemeType.ESSOS,
            # 设置动画,结合学的前端知识
            animation_opts=opts.AnimationOpts(
                animation_delay=1000, #控制动画延时,单位是毫秒
                animation_duration= 200, #控制动画的持续时间
                animation_easing='elasticOut', #控制动画的类型,结合前端中的来理解,比如easeinout,liner这些玩意儿
            )
        )
    )
    .add_xaxis(a)
    .add_yaxis('类别',b)
    # 设置全局配置
    .set_global_opts(
    #     标题设置
        title_opts= opts.TitleOpts(
            title='某地区饮料销量',
            subtitle='万瓶'
        ),
    #      图例设置
        legend_opts=opts.LegendOpts(
            type_='scroll',
            legend_icon='circle'
        ),
        
    #    工具箱设置
        toolbox_opts=opts.ToolboxOpts(),
        # 提示框设置
        tooltip_opts=opts.TooltipOpts(
            background_color='orange',
            textstyle_opts=opts.TextStyleOpts(
                color='white'
            )
        ),
    #     # 可视化映射设置
        visualmap_opts=opts.VisualMapOpts(
            range_text=['超平均线','低于平均线'],
            # range_opacity=0.7,
        ),
    #     # 数据缩放设置
        datazoom_opts=opts.DataZoomOpts(
        ),
    )
    # 设置系列配置
    .set_series_opts(
        label_opts=opts.LabelOpts(
            color='pink',
            position='right'
        ),
        markpoint_opts=opts.MarkPointOpts(
           data=[
               opts.MarkPointItem(
                   type_='max',
                   name='最大值'
               )
           ]
        ),
        markline_opts=opts.MarkLineOpts(
            data=[
                opts.MarkLineItem(
                    type_='average',
                    name='平均值',
                )
            ],
            label_opts=opts.LabelOpts(
                color='red',
            ),
            linestyle_opts=opts.LineStyleOpts(
                color='red',
                type_='dashed'
            )
        )
    )
)
```

```python
c.load_javascript()
```

```python
c.render_notebook()
```

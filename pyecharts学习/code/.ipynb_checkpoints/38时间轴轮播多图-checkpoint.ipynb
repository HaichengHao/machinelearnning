# 38时间轴轮播多图

可以通过时间轮播的方式展示多组图片,可以理解为前端中的轮播图,其实本质就差不多,pyecharts只是给了一个这样的接口

```python
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
import pyecharts.options as opts
from pyecharts.faker import Faker
from pyecharts.charts import Bar,Timeline,Pie
```

### 数据准备

```python
x = Faker.clothes
```

#### 创建一个时间线对象

```python
tl = Timeline()
```

```python
for i in range(2020,2025): #创建5个bar
    bar = (
        Bar()
        .add_xaxis(x)
        .add_yaxis('A',Faker.values(),label_opts=opts.LabelOpts(position='inside'))
        .add_yaxis('B',Faker.values())
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title=f"某商家第{i}年营业额"
            )
        )
    )
    tl.add(bar,f"{i}年") #调用TimeLine()中的.add()方法添加时间线对象
```

```python
tl.load_javascript()
```

```python
tl.render_notebook()
```

```python
tl2 = Timeline()
for i in range(2020,2025): #创建5个bar
     pie = (
        Pie()
        # .add_xaxis(x)
        # .add_yaxis('A',Faker.values(),label_opts=opts.LabelOpts(position='inside'))
        # .add_yaxis('B',Faker.values())
        .add(
            '',
            [list(z) for z in zip(x,Faker.values())],
            radius=['10%','55%'], #radius控制空心圆半径和外径
            center=['50%','50%'],
            rosetype='radius', #控制玫瑰图的类型
        )
     )
     tl2.add(pie,f"{i}年")
        # .set_global_opts(
        #     title_opts=opts.TitleOpts(
        #         title=f"某商家第{i}年营业额"
        #     )
        # )
    # tl2.add(pie,f"{i}年") #调用TimeLine()中的.add()方法添加时间线对象
    
```

## 来一个既有柱状图又有饼状图的,当然,只是作为实验,真正画图不要这样

```python
tl3 = Timeline()
for i in range(2020,2025):
    if i <= 2022:
        bar=(
            Bar()
            .add_xaxis(x)
            .add_yaxis('A',Faker.values(),label_opts=opts.LabelOpts())
            .add_yaxis('B',Faker.values())
        )
        tl3.add(bar,f"{i}年")
    else:
        pie=(
            Pie()
            .add(
                '',
                [list(z) for z in zip(x,Faker.values())],
                radius=['10%','55%'], #radius控制空心圆半径和外径
                center=['50%','50%'], #第一个数值指定的是在整张图横向的位置,第二个是纵向位置
                rosetype='radius', #控制玫瑰图的类型
            )
        )
        tl3.add(pie,f"{i}年")
```

```python
tl3.load_javascript()
```

```python
tl3.render_notebook()
```

# 地理坐标图

```python
from pyecharts.globals import CurrentConfig,NotebookType,ThemeType,ChartType #注意这里新加了一个引入的ChartType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
import pyecharts.options as opts
from pyecharts.faker import Faker
from pyecharts.charts import Geo
```

### 地理坐标图和之前的地图实现效果其实差别不大

但是写法是有区别的,其中需要关注一点<span class='imp'>需要写上.add_schema(maptype='自己指定类型,不清楚就看上一节')</span>

```python
c = (
    Geo()
    .add_schema(maptype='china')
    .add(
        'geo',
        [list(i) for i in zip(Faker.provinces,Faker.values())],
        type_=ChartType.EFFECT_SCATTER, #这里使用的是涟漪散点图
        symbol='triangle',
        effect_opts=opts.EffectOpts(
            color='blue',
            scale=10, #设置涟漪范围大小
            period=10 #设置涟漪速度,
        )
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False)
        
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='地理坐标图_涟漪散点图',
        ),
        visualmap_opts=opts.VisualMapOpts()
    )
)
```

```python
c.load_javascript()
```

```python
c.render_notebook()
```

```python
c2 = (
    Geo()
    .add_schema(maptype='china')
    .add(
        'geo',
        [list(i) for i in zip(Faker.provinces,Faker.values())],
        type_=ChartType.HEATMAP, #这里使用的是涟漪散点图
        symbol='triangle',
        effect_opts=opts.EffectOpts(
            color='blue',
            scale=10, #设置涟漪范围大小
            period=10 #设置涟漪速度,
        )
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False)
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='地理坐标图_热力图',
        ),
        visualmap_opts=opts.VisualMapOpts()
    )
)
```

```python
c2.render_notebook()
```

# 其它绘图

## 带有动态线条的地图

```python
mylst=[list(i) for i in zip(Faker.provinces,Faker.values())]
mylst
```

```python
c3 = (
    Geo(
        init_opts=opts.InitOpts(
            theme=ThemeType.DARK,
            # bg_color='blue'
            
        )
    )
    .add_schema(maptype='china')
    .add(
        'geo',
        [['河南', 39],
         ['山东', 139],
         ['河北', 23],
         ['甘肃', 55],
         ['宁夏', 111],
         ['浙江', 64],
         ['重庆', 68]],
        type_=ChartType.EFFECT_SCATTER, #这里使用的是涟漪散点图
        symbol='triangle',
        effect_opts=opts.EffectOpts(
            color='blue',
            scale=10, #设置涟漪范围大小
            period=10 #设置涟漪速度,
        )
    )
    .add(
        '动态线条',
        [('河南','山东'),('山东','宁夏'),('宁夏','河北'),('河南','浙江'),('重庆','甘肃')],
        type_=ChartType.LINES,
        linestyle_opts=opts.LineStyleOpts(
            color='red',
            curve=0.2 #设置成带有弯曲的曲线
            
        ),
        effect_opts=opts.EffectOpts(
            # symbol='circle',
            color='blue',
        )
        
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False)
        
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='地理坐标图_涟漪散点图',
        ),
        visualmap_opts=opts.VisualMapOpts(
        )
    )
)
```

```python
c3.load_javascript()
```

```python
c3.render_notebook()
```

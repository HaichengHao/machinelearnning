# 面积图

面积图其实就是折线图和x轴之间有一个填充

只需要在添加y轴时添加一个areastyle_opts参数并指定值即可

```python
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts.faker import Faker
from pyecharts.commons.utils import JsCode
import pyecharts.options as opts
from pyecharts.charts import Line
```

```python
a = Faker.week
a
```

```python
b = Faker.values()
b
```

```python
c = (
    Line()
    .add_xaxis(xaxis_data=a)
    .add_yaxis(series_name='week',y_axis=b,is_smooth=True,symbol='triangle',
               areastyle_opts=opts.AreaStyleOpts(
                   opacity=0.7,#指定不透明度
               ))
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            name='星期',
            name_location='end',
            type_='category',
            boundary_gap=False
        ),
        yaxis_opts=opts.AxisOpts(
            name='值',
            name_location='end'
            
        ),
        datazoom_opts=opts.DataZoomOpts()
    )
    .set_series_opts()
    
)
```

```python
c.load_javascript()
```

```python
c.render_notebook()
```

# 堆叠面积图

结合堆叠柱状图来理解,其实只需要加上stack就行了

```python
c2 = (
    Line()
    .add_xaxis(a)
    .add_yaxis(series_name='week',y_axis=b,is_smooth=True,symbol='triangle',
               areastyle_opts=opts.AreaStyleOpts(
                   opacity=0.7,#指定不透明度
                   
               ),
              stack='s1')
    .add_yaxis(series_name='week_',y_axis=Faker.values(),is_smooth=True,
              areastyle_opts=opts.AreaStyleOpts(
                  opacity=0.6,
                  color='red'
              ),stack='s1')
     .add_yaxis(series_name='week_2',y_axis=Faker.values(),is_smooth=True,symbol='rect',
              areastyle_opts=opts.AreaStyleOpts(
                  opacity=0.5,
                  color='purple'
              ),stack='s1')
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            name='星期',
            name_location='end',
            type_='category',
            boundary_gap=False, #设置边界间隙值为false可以让面积图和y轴之间没有间隙,更加美观
        ),
        yaxis_opts=opts.AxisOpts(
            name='值',
            name_location='end'
            
        ),
        toolbox_opts=opts.ToolboxOpts(),
        tooltip_opts=opts.TooltipOpts(
            trigger='axis',
        ),
        datazoom_opts=opts.DataZoomOpts()
    )
    .set_series_opts()
    
)
```

```python
c2.render_notebook()
```

<h1 class="stitle">常见图形的绘制</h1>
<ul>
    <li>
<h2>目标</h2>
        掌握常见统计图及其意义
</li>
</ul>

<p>matplotlib能够绘制折线图、散点图、柱状图、直方图、饼图</p>
<p>我们需要知道不同的统计图的意义，以此来决定选择哪种统计图来呈现我们的数据</p>

<div class="info">
    <h1>常见图形种类及其意义</h1>
    <ul>
        <li>折线图:以折线的上升或者下降来表示统计数量的增减变化的统计图<div>特点:能够显示数据的变化趋势，反映事物的变化情况。(变化)</div>
<div id='api'>api:plt.plot(x,y)</div>
</li>
        <li>散点图:用两组数据构成多个坐标点，考察坐标点的分布，判断两变量之间是否存在某种关联或总结坐标点的分布模式<div>特点:判断变量之间是否存在数量关联趋势，展示离群点(分布规律)</div>
<div id="api">api:plt.scatter(x,y)</div></li>
        <li>柱状图:排列在工作表中的列或行中的数据可以绘制到柱状图中<div>特点:绘制连练得数据，能够一眼看出各个数据的大小，比较数据之间的差别(统计/对比)</div>
<div id="api">api:plt.bar(x,y,width,align="center",**kwargs)</div>
<div id="code">
    参数说明:
    x:需要传递的x轴数据
    y:需要传递的y轴数据
    width:柱状图的宽度
    align:每个柱状图的位置对齐方式
    {'center','edge'},optional,default:'center'
    **kwargs:关键字参数，常用的有color,可以指定柱状图的颜色
</div>
</li>
    <li>
    直方图:由一系列高度不等的纵向条纹或线段表示数据分布的情况。一般用横轴表示数据范围，纵轴表示分布情况。
        <div>特点:绘制连续性的数据展示一组或多组数据的分布状况(统计)</div>
        <div id="api">api:plt.hist(x,bins=None)</div>
    <div id="code">
        x:要传递的数据
        bins:组距
        
</div>
</li>
    <li>饼状图:用于表示不同分类的占比情况，通过弧度大小来对比各种分类</li>
    <div>特点:分类数据的占比情况(占比)</div>
    <div id="api">plt.pie(x,labels=,autopct='%1.2f%%',explode,colors)</div>
    <div id="code">
        x:数量，自动算百分比<br>
        labels:每部分组成<br>
        autopct:占比显示指定%1.2f%%<br>
        shadow:此参数用于在饼状图下方绘制阴影<br>
        labeldistance:此参数是绘制饼状图标签的径向距离<br>
        startangle:此参数用来将饼状图的起点从x轴逆时针旋转角度<br>
        radius:饼状图半径<br>
        counterclock:此参数指定分数方向，顺时针或逆时针<br>
        wedgeprops:此参数是传递给构成饼状图的楔形对象的参数的字典<br>
        center:此参数是图标的中心位置<br>
        frame:如果为true,则此参数用于绘制带有图标的轴框架<br>
        rotatelabels:如果为true,则此参数用于将每个标签旋转到相对应切片的位置<br>
        explode:类似数组默认值为无如果非无则是一个len(x)数组用于指定偏移每个楔形块的半径<br>
        colors:每部分颜色<br>
        pctdistance:此参数是为每个饼图切片的中心与autopct生成的文本开头之间的比率<br>
    </div>
    </ul>
</div>

<h3>1散点图的绘制</h3>

```python
# 1 导包
from matplotlib import pyplot as plt
from pylab import mpl
import random
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 解决部分字符无法正常显示的问题
mpl.rcParams["axes.unicode_minus"]=False
```

```python
# 2 数据准备
x_lst = [10,12,11,41,44,2,12,4,55]
y_lst = [9,12,11,21,40,2,12,4,45]

# 3 设置画布
plt.figure(figsize=(10,10),dpi=100)
plt.grid(True,alpha=0.5)
# 4设置标签
plt.xlabel("数据x")
plt.ylabel("数据y")

# 开始绘画
plt.scatter(x_lst,y_lst,color='r')
```

<h2>2绘制柱状图</h2>

```python
# 0.准备数据
# 电影名字
movie_name = ['雷神3：诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴','降魔传','追捕','七十七天','密战','狂兽','其它']
# 横坐标
x = range(len(movie_name))
# 票房数据
y = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]

# 1.创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 2.绘制柱状图
plt.bar(x, y, width=0.5, color=['b','r','g','y','c','m','y','k','c','g','b'])

# 2.1b修改x轴的刻度显示
plt.xticks(x, movie_name)

# 2.2 添加网格显示
plt.grid(linestyle="--", alpha=0.5)

# 2.3 添加标题
plt.title("电影票房收入对比")

# 3.显示图像
plt.show()
```

<h3>3绘制直方图</h3>

```python
# 创建画布
plt.figure(figsize=(20,8),dpi=100)

# 创建直方图
plt.hist(y,bins=None)

# 绘制
plt.show()
```

<h3>4绘制饼状图</h3>

```python
# 创建画布
plt.figure(figsize=(10,10),dpi=100)

# 创建饼状图
plt.pie(y,autopct='%1.2f%%',labels=movie_name,labeldistance=1,colors=['b','r','g','y','b','r','g'],explode=[0,0.2,0.1,0.2,0.1,0.2,0,0,0,0,0],shadow=True)

# explode爆炸效果,默认值为无如果非无则是一个len(x)数组用于指定偏移每个楔形块的半径

# 指定标题
plt.title("电影票房饼状图",fontsize=20)

# 显示
plt.show()
```

<h2>完善原始折线图-给图形添加辅助功能</h2>
<p>为了更好理解所有基础绘图功能，我们通过天气温度变化的绘图来融合所有的基础API使用</p>
<p style="color:red;font-wight:bold">需求:画出某城市11点到12点一小时以内每分钟的气温变化折线图，温度范围在15-18度</p>

<h3>基础图像</h3>

````python
# 导包
import matplotlib.pyplot as plt
import random
# 数据准备
x=range(60)
# 利用uniform(n,m)可以实现在n,m上随机生成一个实数，利用列表生成式可以实现生成60分钟内每一分钟的数据
y=[random.uniform(15,18) for i in x]
# 创建画布
plt.figure(figsize=(20,8),dpi=100)

# 绘制折线图
plt.plot(x,y)

# 显示折线图
plt.show()
````

<h3>实现给图像增加标题等的功能</h3>

<p>添加自定义刻度</p>
<ul>
    <li>plt.xticks(x,**kwargs)<br>
    其中x：要显示的刻度值<br>
    后面传入的关键字参数可以作为坐标轴的尺度</li>
    <li>plt.yticks(y,**kwargs)<br>
    其中y：要显示的刻度值</li>
    
</ul>

<p>为了解决中文不显示的问题我们需要进行以下操作</p>
<ol>
    <li>from <sapn class="co">pylab</sapn> import <sapn>mpl<sapn></li>

<p style="color:red;font-weight:bolder;">注意，有时候会导致部分字符无法正常显示，那就需要更改axes.unicode_minus参数</p>
<p>mpl.rcParams["axes.unicode_minus"]=False</p>

````python
# 导包
from pylab import mpl
import matplotlib.pyplot as plt
import random
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 解决部分字符无法正常显示的问题
mpl.rcParams["axes.unicode_minus"]=False
# 数据准备
x=range(60)
# 利用uniform(n,m)可以实现在n,m上随机生成一个实数，利用列表生成式可以实现生成60分钟内每一分钟的数据
y=[random.uniform(15,18) for i in x]
# # 创建画布
plt.figure(figsize=(20,8),dpi=100)
# 
# 绘制折线图
plt.plot(x,y)
# 添加x,y轴刻度
x_ticks_label=["11:{}分".format(i) for i in x]
# 如果跑出的图出现字体报错，那么要安装字体，这个可以长久性解决问题
# 或者在python脚本中动态设置matplotlibrc
y_ticks = range(40)

# 修改x,y刻度显示
# 利用列表的索引切分[start:stop:step],让x轴刻度为5个5个调
plt.xticks(x[::5],x_ticks_label[::5]) #坐标刻度不可以直接通过字符串修改
plt.yticks(y_ticks[::5])
# 显示折线图
plt.show()
````

````python
# 导包
from pylab import mpl
import matplotlib.pyplot as plt
import random
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 解决部分字符无法正常显示的问题
mpl.rcParams["axes.unicode_minus"]=False
# 数据准备
x=range(60)
# 利用uniform(n,m)可以实现在n,m上随机生成一个实数，利用列表生成式可以实现生成60分钟内每一分钟的数据
y=[random.uniform(15,18) for i in x]
# # 创建画布
plt.figure(figsize=(20,8),dpi=100)
# 添加网格显示
plt.grid(True,linestyle='--',alpha=0.5)
# 
# 绘制折线图
plt.plot(x,y)
# 添加x,y轴刻度
x_ticks_label=["11:{}分".format(i) for i in x]
# 如果跑出的图出现字体报错，那么要安装字体，这个可以长久性解决问题
# 或者在python脚本中动态设置matplotlibrc
y_ticks = range(40)

# 修改x,y刻度显示
# 利用列表的索引切分[start:stop:step],让x轴刻度为5个5个调
plt.xticks(x[::5],x_ticks_label[::5]) #坐标刻度不可以直接通过字符串修改
plt.yticks(y_ticks[::5])
# 显示折线图
plt.show()
````

````python
# 导包
from pylab import mpl
import matplotlib.pyplot as plt
import random
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 解决部分字符无法正常显示的问题
mpl.rcParams["axes.unicode_minus"]=False
# 数据准备
x=range(60)
# 利用uniform(n,m)可以实现在n,m上随机生成一个实数，利用列表生成式可以实现生成60分钟内每一分钟的数据
y=[random.uniform(15,18) for i in x]
# # 创建画布
plt.figure(figsize=(20,8),dpi=100)
# 添加网格显示
plt.grid(True,linestyle='--',alpha=0.5)
# 
# 绘制折线图
plt.plot(x,y)
# 添加x轴,y轴的描述信息及标题
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("中午11点0分到12点之间的温度变化示意图",fontsize=20)
# 添加x,y轴刻度
x_ticks_label=["11:{}分".format(i) for i in x]
# 如果跑出的图出现字体报错，那么要安装字体，这个可以长久性解决问题
# 或者在python脚本中动态设置matplotlibrc
y_ticks = range(40)

# 修改x,y刻度显示
# 利用列表的索引切分[start:stop:step],让x轴刻度为5个5个调
plt.xticks(x[::5],x_ticks_label[::5]) #坐标刻度不可以直接通过字符串修改
plt.yticks(y_ticks[::5])
# 显示折线图
plt.show()
````

````python
# 导包
from pylab import mpl
import matplotlib.pyplot as plt
import random
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 解决部分字符无法正常显示的问题
mpl.rcParams["axes.unicode_minus"]=False
# 数据准备
x=range(60)
# 利用uniform(n,m)可以实现在n,m上随机生成一个实数，利用列表生成式可以实现生成60分钟内每一分钟的数据
y=[random.uniform(15,18) for i in x]
# # 创建画布
plt.figure(figsize=(20,8),dpi=100)
# 添加网格显示
plt.grid(True,linestyle='--',alpha=0.5)
# 
# 绘制折线图
plt.plot(x,y)
# 添加x轴,y轴的描述信息及标题
plt.xlabel("时间",fontsize=20)
plt.ylabel("温度",fontsize=20)
plt.title("中午11点0分到12点之间的温度变化示意图",fontsize=20)
# 添加x,y轴刻度
x_ticks_label=["11:{}分".format(i) for i in x]
# 如果跑出的图出现字体报错，那么要安装字体，这个可以长久性解决问题
# 或者在python脚本中动态设置matplotlibrc
y_ticks = range(40)

# 修改x,y刻度显示
# 利用列表的索引切分[start:stop:step],让x轴刻度为5个5个调
plt.xticks(x[::5],x_ticks_label[::5],fontsize=20) #坐标刻度不可以直接通过字符串修改
plt.yticks(y_ticks[::5],fontsize=20)
# 设置图片保存
plt.savefig("./tempraturedemo1.png")
# 显示折线图
plt.show()
````

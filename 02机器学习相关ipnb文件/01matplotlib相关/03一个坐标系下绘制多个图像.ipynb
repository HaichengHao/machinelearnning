```python
# 导包
from pylab import mpl
import matplotlib.pyplot as plt
import random
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 解决部分字符无法正常显示的问题
mpl.rcParams["axes.unicode_minus"]=False
# 数据准备
x=range(60)

# 利用uniform(n,m)可以实现在n,m上随机生成一个实数，利用列表生成式可以实现生成60分钟内每一分钟的数据
y=[random.uniform(15,18) for i in x]
# 增加一组数据y1
y1=[random.uniform(11,14) for i in x]
# # 创建画布
plt.figure(figsize=(20,8),dpi=100)
# 添加网格显示
plt.grid(True,linestyle='--',alpha=0.5)

# 绘制折线图
plt.plot(x,y)
# 增加绘制第二条折线图
plt.plot(x,y1,color="green",linestyle="--")
# 添加x轴,y轴的描述信息及标题
plt.xlabel("时间",fontsize=20)
plt.ylabel("温度",fontsize=20)
plt.title("中午11点0分到12点之间的温度变化示意图",fontsize=20)
# 添加x,y轴刻度
x_ticks_label=["11:{}分".format(i) for i in x]
# 如果跑出的图出现字体报错，那么要安装字体，这个可以长久性解决问题
# 或者在python脚本中动态设置matplotlibrc
y_ticks = range(40)

# 修改x,y刻度显示
# 利用列表的索引切分[start:stop:step],让x轴刻度为5个5个调
plt.xticks(x[::5],x_ticks_label[::5],fontsize=20) #坐标刻度不可以直接通过字符串修改
plt.yticks(y_ticks[::5],fontsize=20)

# 显示折线图
plt.show()
```

```python
# 导包
from pylab import mpl
import matplotlib.pyplot as plt
import random
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 解决部分字符无法正常显示的问题
mpl.rcParams["axes.unicode_minus"]=False


# 数据准备
x=range(60)
# 利用uniform(n,m)可以实现在n,m上随机生成一个实数，利用列表生成式可以实现生成60分钟内每一分钟的数据
y=[random.uniform(15,18) for i in x]
# 增加一组数据y1
y1=[random.uniform(11,14) for i in x]


# # 创建画布
plt.figure(figsize=(20,8),dpi=100)
# 添加网格显示
plt.grid(True,linestyle='--',alpha=0.5)

# 绘制折线图
plt.plot(x,y,label="上海")

# 增加绘制第二条折线图
# 在plot后加上label="xxx"可以显示图例
plt.plot(x,y1,color="green",linestyle="--",label="北京")

# 如果图例不能完成添加那就写plt.legend(loc="位置")
plt.legend(loc="lower right",fontsize=20)


# 添加x轴,y轴的描述信息及标题
plt.xlabel("时间",fontsize=20)
plt.ylabel("温度",fontsize=20)
plt.title("中午11点0分到12点之间的温度变化示意图",fontsize=20)

# 添加x,y轴刻度
x_ticks_label=["11:{}分".format(i) for i in x]
# 如果跑出的图出现字体报错，那么要安装字体，这个可以长久性解决问题
# 或者在python脚本中动态设置matplotlibrc
y_ticks = range(40)

# 修改x,y刻度显示
# 利用列表的索引切分[start:stop:step],让x轴刻度为5个5个调
plt.xticks(x[::5],x_ticks_label[::5],fontsize=20) #坐标刻度不可以直接通过字符串修改
plt.yticks(y_ticks[::5],fontsize=20)

# 保存图像
# plt.savefig("./demo.png")
# 显示折线图
plt.show()
```

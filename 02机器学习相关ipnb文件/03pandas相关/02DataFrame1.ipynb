<h1>DataFrame</h1>

```python
# 创建DataFrame
import pandas as pd
import numpy as np
pd.DataFrame(data=None,index=None,columns=None)
```

```python
# index:行标签
# columns:列标签
```

```python
# 通过已有数据创建
pd.DataFrame(np.random.randn(2,3)) #randn为从标准正态分布中返回一个或多个索引值
```

```python
score = np.array([[92, 55, 78, 50, 50],
         [71, 76, 50, 48, 96],
         [45, 84, 78, 51, 68],
         [81, 91, 56, 54, 76],
         [86, 66, 77, 67, 95],
         [46, 86, 56, 61, 99],
         [46, 95, 44, 46, 56],
         [80, 50, 45, 65, 57],
         [41, 93, 90, 41, 97],
         [65, 83, 57, 57, 40]])
```

```python
score
```

```python
score_df = pd.DataFrame(score)
score_df
```

```python
#构造索引序列
subjects=['语文','数学','英语','政治','体育']
# 构造列索引序列
stu = ['同学'+str(i+1) for i in range(score_df.shape[0])]
#shape[0]读取的是第一个维度的长度在即行的长度也就是len(subjects)
stu
```

```python
# 添加行索引
data = pd.DataFrame(score,index=stu,columns=subjects) #注意这里的data放的必须是ndarry形式的数组而不是pandas的Series的数组
data
```

```python
# 索引重设
data.reset_index(drop=False)
# 设置新的下标索引
# drop:默认为False,不删除原来的索引,如果为True,删除原来的索引值
```

```python
# 重置索引,drop=True,会把索引删除
data.reset_index(drop=True)
```

<h2>学习目标</h2>

<p>我们的数据大部分存在于文件当中，所以pandas会支持复杂的IO操作，pandas的API支持众多的文件格式，如CSV、SQL、XLS、JSON、HDF5。</p>

<h2>1csv</h2>

<h3>1.1 read_csv</h3>

<p>pandas.read_csv(filepath_or_buffer, sep =',', usecols )</p>

<p>参数：</p>
<ul>
<li>filepath_or_buffer:文件路径</li>
<li>sep :分隔符，默认用","隔开</li>
<li>usecols:指定读取的列名，列表形式</li>
</ul>

```python
data = pd.read_csv('../data/stock_day.csv',usecols=['open','close'])
data.head(4)
```

<h3>1.2to_csv</h3>

<p>DataFrame.to_csv(path_or_buf=None, sep=', ’, columns=None, header=True, index=True, mode='w', encoding=None)
</p>

<h3>参数:</h3>
<ul>
<li>path_or_buf :文件路径</li>
<li>sep :分隔符，默认用","隔开</li>
<li>columns :选择需要的列索引</li>
<li>header :boolean or list of string, default True,是否写进列索引值</li>
<li>index:是否写进行索引</li>
<li>mode:'w'：重写, 'a' 追加</li>
</ul>

```python
# 举例:选取前十行数据保存便于观察数据
data.to_csv('../data/demotocsv.csv',header=True,mode='a')
```

```python
# 查看一下
pd.read_csv('../data/demotocsv.csv').head(3)
```

<h3>3.1read_json()</h3>

<p>pandas.read_json(path_or_buf=None,orient=None,typ='frame',lines=False)
# 将json格式转换成默认的Pandas DataFrame格式</p>

<ul>
<li>orient : string,Indication of expected JSON string format. <ul>
<li>'split' : dict like {index -&gt; [index], columns -&gt; [columns], data -&gt; [values]}<ul>
<li>split 将索引总结到索引，列名到列名，数据到数据。将三部分都分开了</li>
</ul>
</li>
<li><strong>'records' : list like [{column -&gt; value}, ... , {column -&gt; value}]</strong><ul>
<li>records 以<code>columns：values</code>的形式输出</li>
</ul>
</li>
<li>'index' : dict like {index -&gt; {column -&gt; value}}<ul>
<li>index 以<code>index：{columns：values}...</code>的形式输出</li>
</ul>
</li>
<li><strong>'columns' : dict like {column -&gt; {index -&gt; value}}</strong>,默认该格式<ul>
<li>colums 以<code>columns:{index:values}</code>的形式输出</li>
</ul>
</li>
<li>'values' : just the values array<ul>
<li>values 直接输出值</li>
</ul>
</li>
</ul>
</li>
<li>lines : boolean, default False<ul>
<li>按照每行读取json对象</li>
</ul>
</li>
<li>typ : default ‘frame’， 指定转换成的对象类型series或者dataframe</li>
</ul>

```python
pd.read_json?
```

```python
js_data = pd.read_json('../data/Sarcasm_Headlines_Dataset.json',orient='records',lines=True)
js_data.head(5)
```

<h3>3.2 to_json()</h3>

<p>DataFrame.to_json(*path_or_buf=None*, *orient=None*, *lines=False*)</p>

<h4>参数:</h4>
<ul>
<li>将Pandas 对象存储为json格式</li>
<li><em>path_or_buf=None</em>：文件地址</li>
<li>orient:存储的json形式，{‘split’,’records’,’index’,’columns’,’values’}</li>
<li>lines:一个对象存储为一行</li>
</ul>

```python
js_data.to_json('../data/tst.json',orient='records',lines=True)
```

<h1>补充链接数据库</h1>

<p>使用read_sql()函数读取数据</p>

<p>read_sql()函数可以读取整张数据表,又可以执行SQL语句</p>

<p>pip install sqlalchemy<br>
pip install mysql-connector</p>

<p>语法格式如下</p>

<p class='raw'>
    pandas.read_sql(
            sql,
            con,
            index_col=None,
            paramas = None,
            columns=None,
    )
</p>

<p>
    上述函数中常用参数表示的含义如下:
    <ul>
        <li>sql:表示执行的sql语句</li>
        <li>con:接收数据库连接,表示数据库的连接信息</li>
        <li>index_col:默认为None,如果传入一个列表,则表示为层次化索引</li>
        <li>params:传递给执行方法的参数列表,如params={'name':'vale'}</li>
        <li>columns:接收list表示读取数据的列名,默认为None</li>
    </ul>
</p>

<p>如果发现数据中存在空值,则会使用NaN补全</p>

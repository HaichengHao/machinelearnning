<h2>1索引操作</h2>

<h3>1.1直接使用列索引(先列后行)</h3>

<p>获取2018-02-27这天的close的结果</p>

```python
data['close']['2018-02-27'] #与之前不同现在是先列后行的数据，而且不支持用下标索引拿数据
```

<h3>1.2 结合loc、iloc使用索引访问数据可以先行后列</h3>

<p>获取从'2018-02-27':'2018-02-22'，'open'的结果</p>

```python
# 使用loc:只能指定行列索引的名字
data.loc['2018-02-27':'2018-02-22', 'open']
```

```python
# 使用iloc可以通过索引的下标去获取
# 获取前3天数据,前5列的结果
data.iloc[:3, :5]
```

```python
# 获取2018-02-27的low
data.iloc[0,3]
```

<h3>其它取值方式</h3>

```python
data.loc[data.index[0:5],['open','close']]
```

```python
data.iloc[:5,data.columns.get_indexer(['open','close'])]
```

<h2>3排序操作</h2>

<p>排序有两种形式，一种对于索引进行排序，一种对于内容进行排序</p>

<h3>3.1 DataFrame排序</h3>

<li>使用df.sort_values(by=, ascending=)单个键或者多个键进行排序,</li>


<p class='raw'>
参数：
* by：指定排序参考的键
* ascending:默认升序
    * ascending=False:降序
    * ascending=True:升序
</p>

```python
data.sort_values?
```

```python
# 按照开盘价大小进行排序 , 使用ascending指定按照大小排序
data.sort_values(by='open',ascending=True)
```

```python
# 按照多个键排序
data.sort_values(['open','high'])
```

```python
data.sort_index?
```

```python
data.sort_index(ascending=False)
```

<h3>3.2 Series排序</h3>

<li>使用series.sort_values(ascending=True)进行排序
series排序时，只有一列，不需要参数</li>

```python
data['p_change'].sort_values(ascending=True)
```

<li>使用series.sort_index()进行排序
与df一致</li>

```python
data['low'].sort_index()
```

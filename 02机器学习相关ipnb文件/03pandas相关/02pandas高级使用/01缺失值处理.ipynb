<h1>缺失值处理</h1>

<p>目标</p>
<ul>
<li>应用isnull判断是否有缺失数据NaN</li>
<li>应用fillna实现缺失值的填充</li>
<li>应用dropna实现缺失值的删除</li>
<li>应用replace实现数据的替换</li>
</ul>

<h2>1如何处理NaN</h2>

<ul>
<li>获取缺失值的标记方式(NaN或者其他标记方式)</li>
<li>如果缺失值的标记方式是NaN<ul>
<li>判断数据中是否包含NaN：<ul>
<li>pd.isnull(df),</li>
<li>pd.notnull(df)</li>
</ul>
</li>
<li>存在缺失值nan:<ul>
<li>1、删除存在缺失值的:dropna(axis='rows')<ul>
<li>注：不会修改原数据，需要接受返回值</li>
</ul>
</li>
<li>2、替换缺失值:fillna(value, inplace=True)<ul>
<li>value:替换成的值</li>
<li>inplace:True:会修改原数据，False:不替换修改原数据，生成新的对象</li>
</ul>
</li>
</ul>
</li>
</ul>
</li>
<li>如果缺失值没有使用NaN标记，比如使用"？"<ul>
<li>先替换‘?’为np.nan，然后继续处理</li>
</ul>
</li>
</ul>

<h2>2 电影数据的缺失值处理
</h2>

```python
movie = pd.read_csv('../data/IMDB-Movie-Data.csv')
```

```python
movie.head()
```

<h3>2.1 判断缺失值是否存在</h3>

```python
pd.notnull(movie)#判断数据是否为空
```

```python
# 但是只是这样判断看起来还是很麻烦的所以利用之前学过的numpy的通用判断函数来优化
np.all(pd.notnull(movie))
```

<p> ↑ 可以看到是False,说明数据里有空值</p>

```python
# 当然也可以这样
np.any(pd.isnull(movie))
```

<p>↑有缺失值</p>

```python
# 删除缺失值
data = movie.dropna() #因为调用dropna()后产生了新的对象所以为了方便之后的判断我们用新的对象来承接
```

```python
# 再次查看有没有删除成功
np.all(pd.notnull(data))
```

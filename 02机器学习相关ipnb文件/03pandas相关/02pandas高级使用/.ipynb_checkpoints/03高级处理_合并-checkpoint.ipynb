```python
import pandas as pd
```

```python
data = pd.read_csv('../data/stock_day.csv')
```

```python
# 指定分组区间
p_change=data['p_change']
bins = [-100, -7, -5, -3, 0, 3, 5, 7, 100]
p_count=pd.cut(p_change,bins)
```

```python
dummies=pd.get_dummies(p_count,prefix='rise')
```

<h1>1pd.concat实现数据合并</h1>

<ul>
    <li>pd.concat([data1,data2],axis=1)
    <ul>
    <li>按照行或列进行合并,axis=0为列索引,axis=1为行索引</li>        
</ul>
</li>
</ul>

```python
pd.concat([data,dummies],axis=1)
```

<p>pd,merge()</p>

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                        'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                        'key2': ['K0', 'K0', 'K0', 'K0'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

### 结合mysql的知识来理解,on即为链接指定的约束条件

```python
pd.merge(left,right,on=['key1','key2'])
```

# 内连接就像是取交集如果俩表的key都对应那么才连接

```python
pd.merge(left,right,on=['key1','key2'],how='inner')
# 内连接就像是取交集如果俩表的key都对应那么才连接
```

```python
pd.merge(left,right,on=['key1','key2'],how='left')
```

#### 需要关注一个情况,以哪个表连接,就以哪个表为主

#### 要是不符合这个表的匹配约束,就填充为NaN

```python
pd.merge(left,right,on=['key1','key2'],how='right')
```

#### 要是以outer连接,那么就按照全部的并集为主,只要没有匹配项就显示NaN

```python
pd.merge(left,right,how='outer',on=['key1','key2'])
```

# 外连接就像是在取并集

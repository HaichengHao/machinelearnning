```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../data/stock_day.csv')
p_change=data['p_change']
```

```python
p_change
```

<p>要利用的工具<strong>qcut<strong></p>

```python
# 自动分成差不多数量的类别
pd.qcut?
```

```python
pd.qcut(p_change,10)
```

```python
qcut = pd.qcut(p_change,10) #将数据分成10组
qcut.value_counts() #计算分到每个组数据的个数
```

```python
pd.cut? #自定义区间分组
```

```python
# 指定分组区间
bins = [-100, -7, -5, -3, 0, 3, 5, 7, 100]
p_count=pd.cut(p_change,bins)
```

```python
p_count
```

```python
p_count.value_counts()
```

<h1>独热编码</h1>

```python
pd.get_dummies?
```

```python
dummies = pd.get_dummies(p_count,prefix='rise') 
dummies.head()
```

# 实现一个简单的Matplotlib画图 - 以折线图为例

<h3>第一步，导入matplotlib.pyplot</h3>

```python
# matplotlib.pyplot包含了一系列类似于matlab的画图函数
import matplotlib.pyplot as plt
```

<h3>第二步，创建画布 plt.figure(figsize=(x,y),dpi=n)</h3>

```python
plt.figure(figuresize=(),dpi=)
#figuresize指定图的宽高比
# dpi 图像的清晰度
#返回的是dpi对象
```

<h3>第三步,绘制图像plt.plot(x,y)</h3>

<h3>第四步，显示图像 plt.show()</h3>

<h2>折线图绘制与显示</h2>

<p>举例:展示上海一周的天气比如星期一到星期日的天气温度如下</p>

```python
import matplotlib.pyplot as plt

#创建画布
plt.figure(figsize=(10,8),dpi=100)
# 绘制折线图
plt.plot([i for i in range(1,8)],[15,16,13,12,16,13,10])
# 显示图像
plt.show()
```

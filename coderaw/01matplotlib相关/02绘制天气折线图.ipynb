<h2>完善原始折线图-给图形添加辅助功能</h2>
<p>为了更好理解所有基础绘图功能，我们通过天气温度变化的绘图来融合所有的基础API使用</p>
<p style="color:red;font-wight:bold">需求:画出某城市11点到12点一小时以内每分钟的气温变化折线图，温度范围在15-18度</p>

<h3>基础图像</h3>

```python
# 导包
import matplotlib.pyplot as plt
import random
# 数据准备
x=range(60)
# 利用uniform(n,m)可以实现在n,m上随机生成一个实数，利用列表生成式可以实现生成60分钟内每一分钟的数据
y=[random.uniform(15,18) for i in x]
# 创建画布
plt.figure(figsize=(20,8),dpi=100)

# 绘制折线图
plt.plot(x,y)

# 显示折线图
plt.show()
```

<h3>实现给图像增加标题等的功能</h3>

<p>添加自定义刻度</p>
<ul>
    <li>plt.xticks(x,**kwargs)<br>
    其中x：要显示的刻度值<br>
    后面传入的关键字参数可以作为坐标轴的尺度</li>
    <li>plt.yticks(y,**kwargs)<br>
    其中y：要显示的刻度值</li>
    
</ul>

```python
# 导包
import matplotlib.pyplot as plt
import random
# 数据准备
x=range(60)
# 利用uniform(n,m)可以实现在n,m上随机生成一个实数，利用列表生成式可以实现生成60分钟内每一分钟的数据
y=[random.uniform(15,18) for i in x]
# # 创建画布
plt.figure(figsize=(20,8),dpi=100)
# 
# 绘制折线图
plt.plot(x,y)
# 添加x,y轴刻度
x_ticks_label=["11:{}分".format(i) for i in x]
# 如果跑出的图出现字体报错，那么要安装字体，这个可以长久性解决问题
# 或者在python脚本中动态设置matplotlibrc
y_ticks = range(40)

# 修改x,y刻度显示
# 利用列表的索引切分[start:stop:step],让x轴刻度为5个5个调
plt.xticks(x[::5],x_ticks_label[::5]) #坐标刻度不可以直接通过字符串修改
plt.yticks(y_ticks[::5])
# 显示折线图
plt.show()
```
